```python
import csv 
import pandas as pd
from pathlib import Path
```

```python
data_folder = Path.cwd() / "data"
data_folder.mkdir(exist_ok=True)

# input csv file
INPUT_FILE = "combinedAll.csv"

# output csv file
OUTPUT_FILE = INPUT_FILE.replace(".csv", "ExtraFeatures.csv")
```

```python
df = pd.read_csv(data_folder / INPUT_FILE)
df.columns
```

```python
def add_lagging_feature(new_feature_name, feature_to_track, lag_amount):
    """
    Adds a lagging feature to the DataFrame.
    """
    new_feature_name += str(lag_amount)
    time_col = "second"
    # Track a feature over a rolling window of the specified lag amount.
    df[new_feature_name] = df.rolling(window=lag_amount, min_periods=1, on=time_col)[feature_to_track].sum()
    # Need to shift the count *back* by one, to exclude the current second.
    df[new_feature_name] = df[new_feature_name].shift(1).fillna(0)

add_lagging_feature("recentActivity", "occupied", 5 * 60)  # 5 minutes in seconds
add_lagging_feature("recentActivity", "occupied", 60 * 60)  # 1 hour in seconds
```

```python
df.to_csv(data_folder / OUTPUT_FILE, index=False)
df.head(10)
```
